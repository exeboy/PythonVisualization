```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# sin function

pi = np.pi
x = np.linspace(start=-pi, stop=pi, num=50)  # to create a sequence of 50 evenly spaced numbers
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
# we can change the style of plot 

plt.plot(x, y, color='olive', linewidth=1, linestyle='--', marker='^', markersize=10)
plt.show()
```

```python
# read csv file:

df=pd.read_csv('mpg.csv')
df.head()
```

```python
# scatter plot with hwy(gas usage) and displ(motor capacity)

plt.scatter(x=df['displ'], y=df['hwy'], color='red', alpha=.5) # use alpha for transparency
plt.title('displ - hwy')
plt.xlabel('displ')
plt.ylabel('hwy')
plt.show()
```

```python
# use the cyl property for color of data (cars with more cyl s have higher hwy)

plt.scatter(x=df['displ'], y=df['hwy'], c=df['cyl'], alpha=.5)  # use alpha for transparency
plt.title('displ - hwy')
plt.xlabel('displ')
plt.ylabel('hwy')
plt.colorbar()
plt.show()
```

```python
# plot a 3d chart with the help of mplot3d in matplotlib

from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(df['displ'], df['hwy'], df['cyl'], c=df['cyl'])
plt.show()
```

```python
# bar plot

groups = ['G1', 'G2', 'G3', 'G4', 'G5']
scores = [20, 34, 30, 32, 27]
bar_width = 0.3
plt.bar(groups, scores, width=bar_width, color='black')
plt.title('bar chart')
plt.xlabel('Groups')
plt.ylabel('Scores')
plt.show()
```

```python
# show mean of cty and hwy

classes=df['class'].unique()  # unique items of class column
barwidth = 0.3
cty_mean = []
hwy_mean = []

for x in classes:
    cty_mean.append(df[df['class'] == x]['cty'].mean())
    hwy_mean.append(df[df['class'] == x]['hwy'].mean())

index = pd.factorize(classes)[0] + 1
plt.bar(index, cty_mean, barwidth, color='olive', label='mean city miles per gallon')
plt.bar(index + barwidth, hwy_mean, barwidth, color='purple', label='mean highway miles per gallon')   
plt.xticks(index + barwidth / 2, classes)
plt.legend()
plt.show()
```

```python
# box plot 

df.boxplot(column=['cty','hwy'])
plt.show()
```

```python
# box plot sorted by class

df.boxplot(column=['hwy'], by='class')
plt.show()
```

```python
# counter plot is used to plot 3d chart in 2d

xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
plt.contourf(X, Y, Z)
plt.colorbar()
plt.title('Filled Contours Plot')
plt.show()
```

# Seaborn Library

```python
#june, 17, 2021

import seaborn as sns
```

```python
sns.set_style('darkgrid')
```

```python
# histogram diagram
```

```python
df=pd.read_csv('mpg.csv')
sns.distplot(df[df['cyl']==4].hwy)
plt.show()
```

```python
# scatter plot
```

```python
plt.figure(figsize=(7,7))
ax=sns.scatterplot(data=df, x='displ', y='hwy', hue='cyl', size='cyl', palette='Set1') # Set1 is a library for color of diagram
plt.show()
```

```python
# bar plot
```

```python
plt.figure(figsize=(12, 8))
ax = sns.barplot(data=df, x='class', y='cty', palette='Set3', ci=None, estimator=np.mean)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
ax = sns.boxplot(data=df, x='class', y='hwy', palette='Set3', showfliers=True) # use showfilers to avoid outlier data
plt.show()
```

```python
# heatmap
```

```python
plt.figure(figsize=(8,6))
ax = sns.heatmap(df.corr(),annot=True) # use heatmap for correlation matrices 
plt.show()
```
